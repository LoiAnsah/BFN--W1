```python
#import, read file then print its content
#find number of (rows, columns)
print("Data Info\n")
print("HEAD FOR: TRAIN.CSV\n ")
fileI = 'train.csv'
data_frame = pd.read_csv(fileI)
print(data_frame)
```

```python
# DATA INFROMATION
print("\nINFO FOR: GENGER SUMBISSION.CSV\n ")
print(data_frame.info())
```

```python
#dropping rows with missing values (survived, sex or passenger class)
print("\nDropping rows with missing values\n")
cleaned_data = data_frame.dropna(subset=['Survived','Sex', 'Pclass'])
```

```python
#plotting now 
#EDA 1: Survival by gender
print()
print("Survival by Gender\n")
plt.figure(figsize=(8,6))
sns.countplot(x= 'Sex', hue= 'Survived', data=cleaned_data)
plt.xlabel('Sex')
plt.ylabel('Survived')
plt.title('Survival by Gender')
plt.show()
```

```python
print("Survival by Passenger Class")
plt.figure(figsize=(8,6))
sns.countplot(x= 'Pclass', hue= 'Survived', data=cleaned_data)
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.title('Survival by Gender')
plt.show()
```

```python
print("Passenger class by gender")
plt.figure(figsize=(8,6))
sns.countplot(x= 'Sex', hue= 'Pclass', data=cleaned_data)
plt.xlabel('Sex')
plt.ylabel('Pclass')
plt.title('Survival by Gender')
plt.show()
```
